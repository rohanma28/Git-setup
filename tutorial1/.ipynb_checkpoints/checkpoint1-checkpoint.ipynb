<h2> Data Cleanup </h2>

Import `numpy`, `pandas`, and `matplotlib`.

(Feel free to import other libraries!)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
df = pd.read_csv('../data/states_edu.csv')
df.head()
```

You should always familiarize yourself with what each column in the dataframe represents. Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
df.rename({
    'GRADES_PK_G':'PREK_STUDENTS',
    'GRADES_KG_G':'KINDERGARTENERS',
    'GRADES_4_G':'4TH_GRADERS',
    'GRADES_8_G':'8TH_GRADERS',
    'GRADES_12_G':'SENIORS',
    'GRADES_1_8_G':'ELEM_MIDDLE',
    'GRADES_9_12_G':'HIGH_SCHOOLERS',
    'GRADES_ALL_G':'ALL_STUDENTS',
    'ENROLL':'EST_ALL_STUDENTS'
    },
    axis=1,inplace=True)

#Renamed columns to make them more understandable

df.dropna(subset=['AVG_MATH_8_SCORE'], inplace=True)
df.head()

#Since I am analyzing average 8th grade math score, I dropped rows where that data is not available.
```

Let's compare Michigan to Ohio. Which state has the higher average outcome score across all years?

```python
states = df.groupby("STATE")
mi = states.get_group('MICHIGAN')["AVG_MATH_8_SCORE"].mean()
oh = states.get_group('OHIO')["AVG_MATH_8_SCORE"].mean()
if mi > oh:
    print("Michigan", mi)
else:
    print("Ohio", oh)
```

<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df["IEXP_PER_STUDENT"] = df["INSTRUCTION_EXPENDITURE"] / df["ALL_STUDENTS"]
df["TREV_PER_STUDENT"] = df["TOTAL_REVENUE"] / df["ALL_STUDENTS"]
df["IEXP_PERCENT"] = df["INSTRUCTION_EXPENDITURE"] / (df["STATE_REVENUE"] + df["LOCAL_REVENUE"])
df.head()
```

Feature engineering justification: **<BRIEFLY DESCRIBE WHY YOU MADE THE CHANGES THAT YOU DID\>**

I calculated the instruction expenditure per student because the instruction expenditure is the money that is being invested directly into students' education, meaning that it would have more of an impact than other areas of expenditure on a student's 8th grade math score. I also calculated the total revenue of the state per student, as states with wealthier people will pay more money in taxes per capita, and those wealthier people have more money to invest in their children's education such that their math scores would increase. Finally, I calculated the percent of state and local revenue spent on instruction expenditure, as states that spend more money on instruction in schools should see a rise in scores on 8th grade math.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# X = Total Tax Revenue per Student, Instruction Expenditure per Student, and Average 8th Grade Reading Scores
# y = Average 8th Grade Math Scores
X = df[['TREV_PER_STUDENT','IEXP_PER_STUDENT','AVG_READING_8_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
# model = 
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
# R^2
print(model.score(X_test, y_test))
# mean error
print(np.mean(model.predict(X_test)-y_test))
# mean absolute error
print(np.mean(np.abs(model.predict(X_test)-y_test)))
# root mean squared error
print(np.mean((model.predict(X_test)-y_test)**2)**0.5)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'TREV_PER_STUDENT'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('8th Grade Math Scores')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'TREV_PER_STUDENT'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('8th Grade Math Scores')
plt.title("Model Behavior on Testing Set")
```
